# 2) Rectas y líneas

(5 Puntos)


*   Encuentre la ecuación de la recta que pasa por los puntos (3,5) y (7,8)
*   Encuentre la intersección de la recta $y=2x+3$ con la recta $y=-x$




```python
 # PREGUNTA 2: ESCRIBA CODIGO AQUI

 #----Ecuación de la recta por 2 puntos (3,5) y (7,8)----
m1 = [3,5,1] # -> Punto 1
m2 = [7,8,1] # -> Punto 2

ell = np.cross(m1,m2) # <==== recta que pasa por ambos puntos
a   = ell[0]
b   = ell[1]
c   = ell[2]

print('ell      = ',ell)
print('(a,b,c) = ',(a,b,c))
```

```python
 #----Intersección de la recta y=2x+3 con la recta y=-x----

# Definir las rectas en coordenadas homogéneas
ell_1 = [2, -1, 3]  # Recta 1: 2x - y + 3 = 0 (y = 2x + 3)
ell_2 = [1, 1, 0]   # Recta 2: x + y = 0 (y = -x)

  # Recta 1
a1 = ell_1[0]
b1 = ell_1[1]
c1 = ell_1[2]

x1 = np.array([-5,5])
y1 = -(a1*x1+c1)/b1

plt.plot(x1,y1,color='blue')

  # Recta 2
a2   = ell_2[0]
b2   = ell_2[1]
c2   = ell_2[2]

x2 = np.array([-5,5])
y2 = -(a2*x2+c2)/b2

plt.plot(x2,y2,color='green')

# Calcular la intersección usando el producto cruzado
mi = np.cross(ell_1, ell_2)  # Producto cruzado ℓ1 × ℓ2
xi = mi[0]/mi[2]
yi = mi[1]/mi[2]
print('(xi,yi) = ',(xi,yi))


plt.plot(x1,y1,color='blue')
plt.plot(x2,y2,color='green')
plt.plot(xi, yi, marker='o', markerfacecolor='red', markersize=12)

# Gráfica
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
